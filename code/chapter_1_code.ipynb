## Storing Values in Variables

A **variable** is like a box in the computers memory where you can store a single value. To create a variable in Python you need to name the variable and use the assignment operator, `=`, to assign a value to the variable.

```python
var_1 = 5
var_2 = 3

print(var_1 * var_2)
print(var_1 + var_2)
```
