# Flow Control

In this notebook we will learn all about Python's flow control statements.

## Boolean Values

```python
bool_true = True
bool_false = False
cond = (4 == 4)

print(bool_true)
print(bool_false)
print(cond)
```

## Comparison Operators

```python
4 == 4 # Eqaul
4 != 3 # Not Equal
3 > 2 # GT
2 < 3 # LT
3 >= 3 # GTE
5 <= 5 #LTE
```

## Boolean Operators

Boolean operators are used to compare boolean values (i.e. `True` and `False`. They are `and`, `or`, and `not`.

```python
print(True and True)
print(True or False)
print(not True)
```

## Mixing Boolean and Comparison Operators

```python
(4 > 3) & (2 < 3)
```

## Elements of Flow Control

Flow control statements usually start with a part called the **condition** and are followed by a block of code called the **clause**. Conditions always evaluate to a Boolean value and

```python
my_name = "Alex"
if my_name == "Alex":
  print("Hi, Alex.")
else:
  print("Hi, Stranger")
```

```python
x = 0
while(x < 5):
  print(x)
  x += 1
```

You can use a mixture of `continue` and `break` statements to add further contorl to a `while` loop.

```python
while True:
  print("Who are you?")
  name = input()
  if name != "Alex":
    continue
  print("Hello, Alex. What is the password? (It is a fish.)")
  password = input()
  if password == "swordfish":
    break
print("Access Granted.")
```

We can also use `for` loops to iterate through repetitive code.  

```python
for i in range(5):
  print(i)
```
