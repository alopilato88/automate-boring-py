In this notebook we will learn all about Python's flow control statements.

## Boolean Values

```python
bool_true = True
bool_false = False
cond = (4 == 4)

print(bool_true)
print(bool_false)
print(cond)
```

## Comparison Operators

```python
4 == 4 # Eqaul
4 != 3 # Not Equal
3 > 2 # GT
2 < 3 # LT
3 >= 3 # GTE
5 <= 5 #LTE
```

## Boolean Operators

Boolean operators are used to compare boolean values (i.e. `True` and `False`. They are `and`, `or`, and `not`.

```python
print(True and True)
print(True or False)
print(not True)
```
